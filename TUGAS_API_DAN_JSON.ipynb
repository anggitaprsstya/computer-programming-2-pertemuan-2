```python
import requests

# Input nama dari pengguna
nama = input("Masukkan nama: ")

url = f"https://api.genderize.io?name={nama}"
data = requests.get(url).json()

print("\n=== Hasil Prediksi Gender ===")
print(f"Nama: {data['name']}")
print(f"Gender: {data['gender']}")
print(f"Probabilitas: {data['probability']}")
```
